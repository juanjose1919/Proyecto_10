```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg

from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
```

###PCA

```python
X = data.to_numpy()
```

```python
# estandarizar el dataset
lista = []
for i in X.T:
  u = i.mean()
  s = i.std()
  scal = (i - u) / s
  lista.append(scal)

x_scal = np.array(lista).T
```

```python
#construir la matriz de varianzas y covarianzas
cov_x = np.cov(x_scal.T)
cov_x
```

```python
np.linalg.eig(pd.DataFrame(X).corr().to_numpy())
```

```python
# descomposicin
val_p, vec_p = linalg.eig(cov_x)
val_p, vec_p
```

```python
# 3 componentes

val_p = val_p[:3]
vec_p = vec_p[:, :3]
```

```python
# matriz de proyeccin

W = vec_p
```

```python
W
```

```python
# proyectar X en W

pca_p = x_scal @ W
```

```python
pca_p = pd.DataFrame(pca_p, columns=[f'PC{i}' for i in range(1, pca_p.shape[1] + 1)])
pca_p.head()
```

```python
import seaborn as sb
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min

%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
```

```python
kmeans = KMeans(n_clusters=4).fit(X)
centroids = kmeans.cluster_centers_
centroids
```

```python
#kmeans with PCA
# Predicting the clusters
labels = kmeans.predict(X)
# Getting the cluster centers
C = kmeans.cluster_centers_
colores=['blue','cyan','green','red']
asignar=[]
for row in labels:
  asignar.append(colores[row])

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=asignar,s=60)
ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='*', c=colores, s=1000)
```

En el grfico anterior se puede observar cuatro centroides, entre estos el grupo celeste es el ms diferenciado pues presenta mayor distancia con el resto de los datos, para el grupo rojo la agrupacin se vuelve menos clara y se evidencian datos errneamente clasificados que podran pertenecer a otros grupos. Por ltimo los dos grupos restantes, verde y azul oscuro, no presentan distanciamiento significativo entre sus datos y la participacin de grupos se hace menos clara.

```python
#calcular distancia euclidiana
def dis_eucl(a, b):
  return np.sqrt(sum((a - b)**2))
```

```python
#escoger los centroides, k=num_centroides
centroides = {} #centroides
k = 4 #num_centroides
num_iteraciones = 200
for i in range(k):
  centroides[i] = X[np.random.choice(len(X))]

for i in range(num_iteraciones):
  #distancias
  distances = {}
  for pos, dato in enumerate(X):
    distances[pos] = []
    for _ , centroide in centroides.items():
      distances[pos].append(dis_eucl(centroide, dato))

  #asignacin
  puntos_centroides = {}

  for i in range(k):
    puntos_centroides[i] = []

  for dato, dists in distances.items():
    puntos_centroides[dists.index(min(dists))].append(X[dato])

  #nuevos centroides

  for cent, datos in puntos_centroides.items():
    centroides[cent] = np.average(datos, axis=0)
```

```python
x_0 = np.vstack(puntos_centroides[0])
x_1 = np.vstack(puntos_centroides[1])
x_2 = np.vstack(puntos_centroides[2])
x_3 = np.vstack(puntos_centroides[3])
```

```python
#kmenas with PCA
# Predicting the clusters
labels = kmeans.predict(X)
# Getting the cluster centers
C = kmeans.cluster_centers_
colores=['blue','cyan','green','red']
asignar=[]
for row in labels:
  asignar.append(colores[row])

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x_0[:,0],x_0[:,1])
ax.scatter(x_1[:,0],x_1[:,1])
ax.scatter(x_2[:,0],x_2[:,1])
ax.scatter(x_3[:,0],x_3[:,1])
ax.scatter(centroides[0][0],centroides[0][1], marker='x',color='k')
ax.scatter(centroides[1][0],centroides[1][1], marker='x',color='k')
ax.scatter(centroides[2][0],centroides[2][1], marker='x',color='k')
ax.scatter(centroides[3][0],centroides[3][1], marker='x',color='k')
```

En este grfico se presentan cuatro centroides y cuatro grupos diferenciados, los datos del grupo azul se encuentran a mayor distancia de los dems grupos, los tres restantes presentan menor distancia entre ellos pero an as se evidencia una particin clara y una distancia correspondiente entre sus centroides
